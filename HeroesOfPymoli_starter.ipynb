```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)

purchase_data_df.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results

```python
#variable to hold the number of unique itmes
unique_items = purchase_data_df["Item Name"].nunique()

#sanity check
#unique_items

#variable to hold the average price
avg_price = round(purchase_data_df["Price"].mean(), 2)

#sanity check
#avg_price

#variable to hold the number of purchases
purchases = purchase_data_df["Purchase ID"].count()

#sanity check
#purchases

#variable to hold the total revenue
revenue = purchase_data_df["Price"].sum()
#sanity check
#revenue

#create the df with the above values
purchasing_analysis = {"Number of Unique Items": [unique_items],
                       "Average Price":[avg_price],
                       "Number of Purchases":[purchases],
                       "Total Revenue":[revenue]}

purchasing_analysis_df = pd.DataFrame(purchasing_analysis)

#purchasing_analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#sorted by player name to make it easier to loop through and check for multiples of the same player
sorted_by_sn = purchase_data_df.sort_values("SN")

#sanity check
sorted_by_sn
```

```python
male_count = 0
female_count = 0
other_count = 0

#loop to keep a tally of players' genders
for row in range(0, len(sorted_by_sn) - 1):
       
    #sanity check
    #print(sorted_by_sn.iloc[row, 1])
    #print(sorted_by_sn.iloc[row + 1, 3])
          
    if row < len(sorted_by_sn):
        
        #conditional to check for different players
        if sorted_by_sn.iloc[row, 1] != sorted_by_sn.iloc[row + 1, 1]:
        
            #keep count of male players
            if sorted_by_sn.iloc[row, 3] == "Male":
                male_count = male_count + 1
                
        
            #keep count of female players
            elif sorted_by_sn.iloc[row, 3] == "Female":
                female_count = female_count + 1
        
            #keep count of other/non-disclosed players
            else:
                other_count = other_count + 1
                
    elif row == len(sorted_by_sn) + 1:
        
        #keep count of male players
        if sorted_by_sn.iloc[row, 3] == "Male":
            male_count = male_count + 1
                
        
        #keep count of female players
        elif sorted_by_sn.iloc[row, 3] == "Female":
            female_count = female_count + 1
        
        #keep count of other/non-disclosed players
        else:
            other_count = other_count + 1
```

```python
#sanity check
#print(male_count)
#print(female_count)
#print(other_count)
```

```python
#calculations for the gender percentages
percent_male = round((male_count / total_players) * 100, 2)

percent_female = round((female_count / total_players) * 100, 2)

percent_other = round((other_count / total_players) * 100, 2)
```

```python
#variable to hold the gender data
gender_data = {'Total Count':[male_count, female_count, other_count],
               'Percentage of Players':[percent_male, percent_female, percent_other]}

#gender demo dataframe
gender_demo_df = pd.DataFrame(gender_data)

#renaming the indexes
gender_demo_df = gender_demo_df.rename(index={0: "Male", 1:"Female", 2:"Other/Non-Disclosed"})

gender_demo_df["Percentage of Players"] = gender_demo_df["Percentage of Players"].map('{:,.2f}%'.format)

gender_demo_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#variables to hold the total counts for all genders
all_male_count = 0
all_female_count = 0
all_other_count = 0

#variables to hold the total purchases
male_purchase_totals = 0
female_purchase_totals = 0
other_purchase_totals = 0
```

```python
#loop to keep a tally of players' genders
for row in range(0, len(purchase_data_df)):
    
    #keep count of male players and purcahses
    if purchase_data_df.iloc[row, 3] == "Male":
        all_male_count = all_male_count + 1
        
        #keeping track of total purchases
        male_purchase_totals = round(male_purchase_totals + purchase_data_df.iloc[row, 6], 2)
        
    #keep count of female players and purchases
    elif purchase_data_df.iloc[row, 3] == "Female":
        all_female_count = all_female_count + 1
        
        #keeping track of total purchases
        female_purchase_totals = round(female_purchase_totals + purchase_data_df.iloc[row, 6], 2)
        
    #keep count of other/non-disclosed players and purchases
    else:
        all_other_count = all_other_count + 1
        
        #keeping track of total purchases
        other_purchase_totals = round(other_purchase_totals + purchase_data_df.iloc[row, 6], 2)
    
```

```python
#calculating the average purchase price
avg_male_purchase = round(male_purchase_totals / all_male_count, 2)

avg_female_purchase = round(female_purchase_totals / all_female_count, 2)

avg_others_purchase = round(other_purchase_totals / all_other_count, 2)
```

```python
#calculating the average total purchase per person
avg_per_male = round(male_purchase_totals / male_count, 2)

avg_per_female = round(female_purchase_totals / female_count, 2)

avg_per_other = round(other_purchase_totals / other_count, 2)

```

```python
gender_purchasing_analysis = {"Purchase Count":[all_female_count, all_male_count, all_other_count],
                              "Average Purchase Price":[avg_female_purchase, avg_male_purchase, avg_others_purchase],
                              "Total Purchase Value":[female_purchase_totals, male_purchase_totals, other_purchase_totals],
                              "Avg Total Purchase per Person":[avg_per_female, avg_per_male, avg_per_other]}
```

```python
#creating the df for the gender purchasing analysis
gender_purchasing_analysis_df = pd.DataFrame(gender_purchasing_analysis)

gender_purchasing_analysis_df = gender_purchasing_analysis_df.rename(index = {0:"Female", 1:"Male", 2:"Other/Non-Disclosed"})

gender_purchasing_analysis_df["Average Purchase Price"] = gender_purchasing_analysis_df["Average Purchase Price"].map('${:,.2f}'.format)
gender_purchasing_analysis_df["Total Purchase Value"] = gender_purchasing_analysis_df["Total Purchase Value"].map('${:,.2f}'.format)
gender_purchasing_analysis_df["Avg Total Purchase per Person"] = gender_purchasing_analysis_df["Avg Total Purchase per Person"].map('${:,.2f}'.format)
gender_purchasing_analysis_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()

```python
#check to see the max age
#purchase_data_df["Age"].max()
```

```python
#creating the bins to cut the data into age groups
bins = [0, 9, 14, 19, 24, 29, 34, 39, 45]
```

```python
#setting the labels for the age groups
labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
#new df for just the age and SN
age_df = purchase_data_df[["SN", "Age"]]

#new name for after the duplicates are dropped
no_duplicates_df = age_df.drop_duplicates("SN")

#sanity check
#no_duplicates_df
```

```python
#creating the cutoffs
no_duplicates_df["Age Groups"] = pd.cut(no_duplicates_df["Age"], bins, labels=labels)

just_age_df = no_duplicates_df.drop(["SN"], axis = 1)
```

```python
#grouping by age_groups
age_groups_df = just_age_df.groupby("Age Groups")

age_calc_df = age_groups_df.count()

age_calc_df
```

```python
#calculating the percentages
under_10 = (age_calc_df.iloc[0]["Age"] / total_players) * 100
ten_fourteen = (age_calc_df.iloc[1]["Age"] / total_players) * 100
fifteen_nineteen = (age_calc_df.iloc[2]["Age"] / total_players) * 100
twenty_twentyfour = (age_calc_df.iloc[3]["Age"] / total_players) * 100
twentyfive_twentynine = (age_calc_df.iloc[4]["Age"] / total_players) * 100
thirty_thirtyfour = (age_calc_df.iloc[5]["Age"] / total_players) * 100
thirtyfive_thirtynine = (age_calc_df.iloc[6]["Age"] / total_players) * 100
fourty_over = (age_calc_df.iloc[7]["Age"] / total_players) * 100

#making a list for the new column
percentage_of_players = [under_10, ten_fourteen, fifteen_nineteen, twenty_twentyfour, twentyfive_twentynine, thirty_thirtyfour, thirtyfive_thirtynine, fourty_over]

percentage_of_players
```

```python
#adding the last column for the percentages
age_calc_df["Percentage of Players"] = percentage_of_players
```

```python
#sanity check
#age_calc_df
```

```python
#reformatting the percetnages
age_calc_df["Percentage of Players"] = age_calc_df["Percentage of Players"].map('{:,.2f}%'.format)

#renaming the age column to total count
age_calc_df = age_calc_df.rename(columns={"Age Groups": "", "Age":"Total Count"})

age_calc_df
```
